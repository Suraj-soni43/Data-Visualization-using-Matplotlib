# Installing the matplotlib

```python
# importing the package library in code
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

# line graph (plot)

```python
x = [1,2,3]
y = [6,4,5]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [6,4,5]                       #.plot() display a graph in line format as defoult
plt.plot(x,y,'r',label='Line 1')  #'r' to color the line in Red color
plt.title('Information')          #.title() for asinging a title to a graphplt.xlabel('x-label')
plt.xlabel('x-label')             #.xlabel() used to display a name X-axis
plt.ylabel('y-label')             #.ylabel() used to display a name Y-axis        
plt.legend()
plt.show()                        #.show() to display a graph 
```

```python
#Line graph with Grid
x = [1,2,3]
y = [6,4,5]
plt.plot(x,y,'r',label='LINE 1')
x2 = [1,2,3]
y2 = [8,9,1]
plt.plot(x2,y2,'b',label='Line 2')
plt.title('Information')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True,color='y')
plt.show()
```

# Bar graph

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,]
y=[3,12,3,14,53]
plt.bar(x,y,label='Example',color='g')
plt.title('Information')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Histogram (hist)

```python
ran = [0,10,20,30,40,50]
values = [1,2,4,3,32,1,11,12,33,2,44,21,15,31,23,45,4,33]
plt.hist(values,ran, label='Histogram Example', color='k')
plt.title('Information')
plt.xlabel('X-axis')
plt.ylabel('Y-axia')
plt.legend()
plt.show()
```

# Scatter plot (scatter)

```python
x = [1,2,3,4,5,6,7,8]
y = [2,12,3,4,51,12,3,14]
plt.scatter(x,y,label='skitskat',color='k')
plt.title('Information')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Stack/Area plot (stackplot)

```python
days = [1,2,3,4,5]
sleep = [2,4,6,2,4]
code = [2,5,1,4,10]
work = [3,4,2,1,10]
plt.stackplot(days,sleep,code,work,labels=['sleep','code','work'],colors=['g','r','b'])
plt.title('Stack Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Pie chart (pie)

```python
activities = ['sleep','code','work','eat']
slices = [7,2,13,2]
plt.pie(slices,labels = activities,colors=['r','g','b','y'],startangle=90,shadow=True,explode=(0.1,0,0,0),autopct='%1.1f%%')

plt.title('Pie Chart of activities')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# Subplot

```python
x = [1,2,3]
y = [6,4,5]
x2 = [1,2,3]
y2 = [8,9,1]
plt.subplot(211)
plt.plot(x,y,'r',label='LINE 1')
plt.subplot(212)
plt.plot(x,y,'k',label='Line 2')
plt.show()
```

```python
import numpy as np
x = np.arange(0,np.pi*3,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
import covid
import matplotlib.pyplot as plt
```

```python
cov = covid.Covid()
name = input('ENTER THE NAME OF THE COUNTRY ')
virusdata = cov.get_status_by_country_name(name)

remove =['id', 'country', 'latitude', 'longitude', 'last_update']
for i in remove:
    virusdata.pop(i)
    
total = virusdata.pop('confirmed')
slices = virusdata.values()
keys = virusdata.keys()
plt.pie(slices,labels = keys,colors=['r','g','y'],startangle=270,shadow=True,explode=(0,0.5,0),autopct='%1.1f%%')
plt.title(name+'\n'+'Total Cases:'+str(total))
plt.legend()
plt.show()
```
